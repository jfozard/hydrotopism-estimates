# Hydrotropism - analysing sorbitol gradients in agar gels

Measurements of sucrose diffusion in c. 1% agar are in the range $.5 \times 10^6$ to $1 \times 10^5 \mbox{cm}^2 s^{-1}$. [Refs http://www.sciencedirect.com/science/article/pii/0141022993900546 http://pubs.acs.org/doi/abs/10.1021/bi00910a019 ]

Diffusion equation for sorbitol concentration $$ \frac{\partial c}{\partial t}  = D \frac{\partial^2 c}{\partial x^2}$$ Initial conditions $$ c(x,0) = \begin{cases} A & x<0 \\ 0 & x \ge 0 \end{cases}$$
This has solution $$ c(x,t) = A \, \mbox{erfc} \left(\frac{x}{2 \sqrt{D t}} \right) $$

Root lies at approximately $2.5 \, \mbox{mm}$ from the interface, is approximately $100 \, \mu \mbox{m}$ across.

Data from paper (Takahashi et al 2002; http://link.springer.com/article/10.1007%2Fs00425-002-0840-3)

### Reproduce figures from paper

```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
from math import erfc, sqrt
```

```python
times = [0., 2., 4., 24.]
distances = [-6., -4., -2., 2., 4., 6.]
labels = ['a', 'b', 'c', 'd', 'e', 'f']
```

```python
def soln(x, t, A, D):
    if t>0:
        return 0.5*A*erfc(x/(2*sqrt(D*t)))
    else:
        return A*(x<0)
```

```python
A = -1.0
D = 1e-5*(10)**2*3600 # Conversion from cm^2/s to mm^2/hr
```

```python
plt.figure()
s = np.linspace(-6, 6, 100)
for i in range(6):
    plt.plot(s, [soln(y, i, A, D) for y in s])
plt.xlabel("Distance from boundary (mm)")
plt.ylabel("Water potential (MPa)")
plt.title("Evolution of water potential over 6 hours")
plt.draw()
```

```python
# Figure b
plt.figure()
t = np.linspace(0, 24, 1000)
for s in distances:
    plt.plot(t, [soln(s, y, A, D) for y in t])
plt.xlabel("Time (hr)")
plt.ylabel("Water Potential (MPa)")
plt.title("Water potential at points a-f")
plt.draw()
```

```python
# Figure c - in space
plt.figure()
s = np.linspace(-6, 6, 100)
for t in [4.]:
    plt.plot(s, [soln(y, t, A, D) for y in s])
    
plt.xlabel("Distance from boundary (mm)")
plt.ylabel("Water potential (MPa)")
plt.title("Spatial distribution at 4 hours")
plt.draw()
```

### Average value and gradient across a root 2.5 mm away from the border

```python
plt.figure()
s_mid = 2.5
t = np.linspace(0, 12, 100)
plt.plot(t, [1000*soln(s_mid, y, A, D) for y in t])
plt.title("Water potential at root midline $x=2.5$mm")
plt.xlabel("Time (hr)")
plt.ylabel("Water Potential (kPa)")
plt.plot(t, -300*np.ones((100,)))
plt.legend(["outside", "cytosol"])
plt.savefig("average.pdf", format="pdf")

plt.draw()
```

```python
plt.figure()
width = 0.1
s_mid = 2.5
def abs_diff(y):
    return soln(s_mid+width/2, y, A, D) - soln(s_mid-width/2, y, A, D)
plt.plot(t, [1000*abs_diff(y) for y in t])
plt.title("Absolute difference in  water potential across root diameter")
plt.xlabel("Time (hr)")
plt.ylabel("Difference in water Potential (kPa)")
plt.savefig("absolute.pdf", format="pdf")
plt.draw()
```

```python
plt.figure()
width = 0.1
s_mid = 10
def abs_diff(y):
    return soln(s_mid+width/2, y, A, D) - soln(s_mid-width/2, y, A, D)
plt.plot(t, [1000*abs_diff(y) for y in t])
plt.title("Absolute difference in  water potential across root (fixed at $s=10$ mm from boundary)")
plt.xlabel("Time (hr)")
plt.ylabel("Difference in water Potential (KPa)")
plt.draw()
```

###### Note that these absolute differences are fairly small compared to the c. 300 kPa turgor pressure within the cells

```python
plt.figure()
smid = 2.5
def rel_diff(y):
    w1 = soln(s_mid+width/2, y, A, D)
    w2 = soln(s_mid-width/2, y, A, D)
    return (w2 - w1)/(0.5*(w1+w2))
    #return w2

plt.plot(t[10:40], [rel_diff(y) for y in t[10:40]])
plt.title("Difference in  water potential across root relative to the average water potential outside root")
plt.xlabel("Time (hr)")
plt.ylabel("Relative difference")
plt.savefig("relative.svg", format="svg")
plt.draw()
```
